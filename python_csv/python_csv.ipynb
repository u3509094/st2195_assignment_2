```python
from bs4 import BeautifulSoup
```

```python
with open("Comma-separated values - Wikipedia.html", encoding = "utf8") as f:
    doc = BeautifulSoup(f, "html.parser")

print(doc)
```

```python
csv_html = doc.prettify()
print(csv_html)
```

```python
car_csv = doc.find_all("pre")
print(car_csv)
```

```python
print(car_csv[-3])
```

```python
car_table_csv = car_csv[-3].string
print(type(car_table_csv))
print(car_table_csv)
```

```python
import csv
with open('car.csv', 'w') as car_csv:
    writer = csv.writer(car_csv)

    writer.writerows(csv.reader((str(car_table_csv).splitlines())))
```
